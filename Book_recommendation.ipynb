```python
book1_df=books_df.pivot_table(index='UserID',
                        columns='BookTitle',
                       values='BookRating').reset_index(drop=True)
```

```python
book1_df
```

```python
book1_df.index=books_df.UserID.unique()
book1_df
```

```python
book1_df.fillna(0,inplace=True)
book1_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book_sim=1-pairwise_distances(book1_df.values,metric='cosine')
```

```python
book_sim
```

```python
book_sim_df=pd.DataFrame(book_sim)
```

```python
book_sim_df.index=books_df.UserID.unique()
book_sim_df.columns=books_df.UserID.unique()
```
